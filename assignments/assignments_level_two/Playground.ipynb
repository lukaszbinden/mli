```python
x1 = 2
x2 = 4
```

```python
def f():
    print(f"f() x1: {x1}")
    print(f"f() x2: {x2}")
    x1r = x1 + 2
    x2r = x2 + 4
    return x1r + x2r
    
```

```python
def g():
    global x1, x2
    x1 = x1 + 1
    x2 = x2 + 1
```

```python
print(x1)
print(x2)
res = f()
print("f():", res)
print(x1)
print(x2)
print("g()..")
g()
res = f()
print("f():", res)
print(x1)
print(x2)
```
